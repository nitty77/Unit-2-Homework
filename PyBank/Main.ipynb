```python
# import 

import os
import csv

#path for file
budget_data_csv = os.path.join(r"C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New Folder\budget_data.csv")

#output
text_path = "text.txt"

#variables
total_mon = 0
total_rev = 0
rev = []
prev_rev = 0
mon_of_change = []
rev_change = 0
greatest_dec = ["", 9999999]
greatest_inc = ["", 0]
rev_change_list = []
rev_avg = 0

with open(r'C:\Users\aarora\OneDrive - AMERIBEST HOMECARE\Desktop\New Folder\budget_data.csv') as csvfile:  
    csvreader = csv.DictReader(csvfile)
    
    for row in csvreader:
        total_mon += 1
        
        #Net total amount
        total_rev = total_rev + int(row["Profit/Losses"])

        #Average Revenue
        rev_change = float(row["Profit/Losses"])- prev_rev
        prev_rev = float(row["Profit/Losses"])
        rev_change_list = rev_change_list + [rev_change]
        mon_of_change = [mon_of_change] + [row["Date"]]
       

        #Greatest increase in revenue 
        if rev_change>greatest_inc[1]:
            greatest_inc[1]= rev_change
            greatest_inc[0] = row['Date']

        #Greatest decrease in revenue 
        if rev_change<greatest_dec[1]:
            greatest_dec[1]= rev_change
            greatest_dec[0] = row['Date']
    rev_avg = sum(rev_change_list)/len(rev_change_list)
    
    with open(text_path, 'w') as file:
    
        file.write("Python Unit 2 Homework\n")
        file.write("Total Months: %d\n" % total_mon)
        file.write(" Net Total Revenue: $%d\n" % total_rev)
        file.write("Average Revenue Change $%d\n" % rev_avg)
        file.write("Greatest Increase in Revenue: %s ($%s)\n" % (greatest_inc[0], greatest_inc[1]))
        file.write("Greatest Decrease in Revenue: %s ($%s)\n" % (greatest_dec[0], greatest_dec[1]))

```
